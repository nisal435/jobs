```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset into a Pandas DataFrame
file_path = "/Users/nisalgunawardhana/Desktop/jobs_in_data.csv"  
jobs_df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset to understand its structure
jobs_df.head()
```

```python
# Remove duplicates
jobs_df = jobs_df.drop_duplicates()
```

```python
# Data Analysis

# Calculate the average salary by job category
average_salary_by_category = jobs_df.groupby('job_category')['salary_in_usd'].mean()
```

```python
# Data Visualization

# average salary by job category
plt.figure(figsize=(10, 6))
sns.barplot(x=average_salary_by_category.index, y=average_salary_by_category.values)
plt.title('Average Salary by Job Category')
plt.xlabel('Job Category')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Salary Distribution

plt.figure(figsize=(10, 6))
sns.histplot(jobs_df['salary_in_usd'], bins=30, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Salaries by Experience Level

plt.figure(figsize=(12, 8))
sns.boxplot(x='experience_level', y='salary_in_usd', data=jobs_df)
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```

```python
# Job Category Countplot

plt.figure(figsize=(12, 6))
sns.countplot(x='job_category', data=jobs_df, order=jobs_df['job_category'].value_counts().index)
plt.title('Job Category Distribution')
plt.xlabel('Job Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Salary vs. Experience Level

plt.figure(figsize=(12, 8))
sns.scatterplot(x='experience_level', y='salary_in_usd', hue='employment_type', data=jobs_df)
plt.title('Salary vs. Experience Level by Employment Type')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.legend(title='Employment Type')
plt.show()
```

```python
# Violin plot of salary distribution by job category with legend
plt.figure(figsize=(14, 8))
sns.violinplot(x='job_category', y='salary_in_usd', data=jobs_df, palette='viridis')

# Add legend
plt.legend(title='Salary Distribution', loc='upper right', labels=['Low', 'Medium', 'High'])

plt.title('Salary Distribution by Job Category')
plt.xlabel('Job Category')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=45)
plt.show()
```
